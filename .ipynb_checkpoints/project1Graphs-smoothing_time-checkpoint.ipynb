# Total Footfall Among Top 10 MTA Stations
Timeframe: January to June 2019

```python
with open('top10station_janjun2019','rb') as read_file:
    top10station_janjun2019 = pickle.load(read_file)
```

```python
plt.title('Total Footfall Among Top 10 MTA Stations\n January to June 2019', weight='bold',family='sans-serif',fontsize = 20 ,loc='center',color=mtaLogoColor)

plt.xlabel('MTA Stations', fontsize = 10,color=mtaLogoColor)
plt.ylabel('Commuters',fontsize = 10,color=mtaLogoColor)

plt.ticklabel_format(style='plain')
plt.xticks(rotation=45, ha="right")

# Used color to highlight top 3 instead
plt.bar(top10station_janjun2019['STATION'][:10], top10station_janjun2019['TOTAL_TRAFFIC'][:10], 
        color = [mtaColor[0], mtaColor[0], mtaColor[0],
                 mtaColor[1], mtaColor[1], mtaColor[1],
                 mtaColor[1], mtaColor[1], mtaColor[1],
                 mtaColor[1]])
plt.savefig('total_footfall_top_10_stations_jan_to_jun_2019_highlighted.png',bbox_inches = "tight");

# All gray for transition purposes
plt.bar(top10station_janjun2019['STATION'][:10], top10station_janjun2019['TOTAL_TRAFFIC'][:10], 
        color = mtaColor[1])

plt.savefig('total_footfall_top_10_stations_jan_to_jun_2019_gray.png',bbox_inches = "tight");

# Presentation Graph ---------------------------------------------
# MTA line colors is thoughtful, but is too varied and has no explicit purpose
plt.bar(top10station_janjun2019['STATION'][:10], top10station_janjun2019['TOTAL_TRAFFIC'][:10], 
        color = mtaLineColor)
plt.savefig('total_footfall_top_10_stations_jan_to_jun_2019_colored.png',bbox_inches = "tight");
```

# Total Weekday and Weekend Footfall Among Top 10 MTA Stations
Timeframe: January to June 2019

```python
with open('top10weekdayAndWeekend_total','rb') as read_file:
    top10weekdayAndWeekend_total = pickle.load(read_file)
```

```python
# top10weekdayAndWeekend_total
```

```python
plt.title('Total Weekday and Weekend Footfall Among Top 10 MTA Stations\n January to June 2019', weight='bold',family='sans-serif',fontsize = 20 ,loc='center',color=mtaLogoColor)

plt.xlabel('MTA Stations', fontsize = 10,color=mtaLogoColor)
plt.ylabel('Commuters',fontsize = 10,color=mtaLogoColor)

plt.ticklabel_format(style='plain')
plt.xticks(rotation=45, ha="right")

x = top10weekdayAndWeekend_total.loc[top10weekdayAndWeekend_total['WEEKFRAME'] == 'WEEK'].sort_values(by='STATION')
y = top10weekdayAndWeekend_total.loc[top10weekdayAndWeekend_total['WEEKFRAME'] == 'WEEKEND'].sort_values(by='STATION')

# print(x)
# print(y)

x_height = x['TOTAL_TRAFFIC']
y_height = y['TOTAL_TRAFFIC']

# print(x_height)
# print(y_height)

N = 10
ind = np.arange(N)  # the x locations for the groups
width = 0.4       # the width of the bars

station_labels = ('125 ST', '14 ST-UNION SQ', '23 ST', '34 ST-HERALD SQ', '34 ST-PENN STA', '42 ST-PORT AUTH', '86 ST', 'FULTON ST', 'GRD CNTRL-42 ST','TIMES SQ-42 ST')
plt.xticks(ind + width / 2, station_labels)

plt.bar(ind, x_height, width = width, color = mtaColor[0], label = 'Weekday')
plt.bar(ind+width, y_height,width = width, color = mtaColor[1], label='Weekend')
plt.legend()

plt.savefig('weekday_and_weekend_top10_stations_jan_to_jun_2019_total_fixed.png', bbox_inches = "tight");


# Presentation Graph ---------------------------------------------
plt.bar(ind, x_height, width = width, color = mtaLineColor, label = 'Weekday')
plt.bar(ind+width, y_height,width = width, color = mtaLightenedColors, label='Weekend')

plt.savefig('weekday_and_weekend_top10_stations_jan_to_jun_2019_total_colored.png', bbox_inches = "tight");
```

# 24H Footfall for 34 St - Penn Station
Timeframe: January to June 2019

```python
with open('janjun2019','rb') as read_file:
    df_filtered = pickle.load(read_file)
```

```python
df_34ST = df_filtered[df_filtered['STATION'] == '34 ST-PENN STA']
df_34ST_time = df_34ST.groupby(['TIME']).mean()
```

```python
df_34ST_time

df_34ST_time.to_csv(r'34st_uncleaned_time.csv')
```

```python
import datetime

plt.title('24H Footfall for 34 St - Penn Station\n Jan to June 2019', weight='bold',family='sans-serif',fontsize = 20 ,loc='center',color=mtaLogoColor)

plt.xlabel('Time', fontsize = 10,color=mtaLogoColor)
plt.ylabel('Commuters',fontsize = 10,color=mtaLogoColor)

plt.ticklabel_format(style='plain')
plt.xticks(rotation=45, ha="right")

test = []
for i in df_34ST_time.index:
    a = i.split(':')
    test.append(datetime.time(int(a[0]),int(a[1]),int(a[2])))
# print(test)

test2 = [datetime.time(x,0,0) for x in range(24)]
# print(df_34ST_time.index)

# time_labels = [str(x) for x in range(24)]
plt.xticks(test2)

plt.plot(test, df_34ST_time['TOTAL_TRAFFIC']);

plt.savefig('24H_34St.png', bbox_inches = "tight");
```

```python
df_testing = df_filtered[df_filtered['STATION'] == '34 ST-PENN STA'].copy()


df_testing['TIME']= pd.to_timedelta(df_testing['TIME'])
df_testing.info()

bins = []
labels = []
for x in range(25):
    bins.append(pd.Timedelta(hours = x, minutes = 0))
    bins.append(pd.Timedelta(hours = x, minutes = 30))
    if x < 10:
        labels.append('0%s:00:00' % x)
        labels.append('0%s:30:00' % x)
    else: 
        labels.append('%s:00:00' % x)
        labels.append('%s:30:00' % x)

labels = labels[:-1]
df_testing['TIMEBINS'] = pd.cut(df_testing['TIME'], bins, labels = labels)

# df_testing.describe()
df_testing = df_testing.groupby(['TIMEBINS']).mean()
# df_testing.fillna(0)
# print(len(labels))
# print(len(bins))
# print(labels)
# print(bins)
# resampled_time = df_testing(['TIME']).resample('30min').mean()
# resampled_time
# df_testing = df_testing.groupby(['TIME']).resample('30min').mean().copy()
# df_testing = df_testing.groupby(['TIME']).mean() # .resample('2T')
```
