```python
df_station = df_filter_months.groupby(['STATION','TIMESTAMP','DAY_OF_WEEK','TIMEFRAME','WEEKFRAME','DATE'],sort = True).sum().reset_index()
```

```python
df_station[(df_station['STATION'] == '34 ST-PENN STA') & (df_station['DATE'] == '04/19/2019')]
```

```python
df_station[(df_station['TOTAL_TRAFFIC'] > 500000)]
```

```python
df_perStationPerTimeFrameMean = df_station.groupby(['STATION','DAY_OF_WEEK','TIMEFRAME','WEEKFRAME'], sort=True).mean().reset_index()
df_perStationPerTimeFrameMean = df_perStationPerTimeFrameMean.sort_values(by = ['TOTAL_TRAFFIC'],ascending=[False])
print(df_perStationPerTimeFrameMean[:25])
```

```python
df_perStationPerTimeFrameMean = df_station.groupby(['STATION','DAY_OF_WEEK','TIMEFRAME','WEEKFRAME'], sort=True).mean().reset_index()
df_perStationPerTimeFrameMean = df_perStationPerTimeFrameMean.sort_values(by = ['TOTAL_TRAFFIC'],ascending=[False])
print(df_perStationPerTimeFrameMean[df_perStationPerTimeFrameMean['TIMEFRAME'] == 'MORNING'][:25])
```

```python

a_dims = (11.7, 8.27)
data = df_perStationPerTimeFrameMean[df_perStationPerTimeFrameMean['TIMEFRAME'] == 'MORNING'][:25]
sns.set_style('ticks')
fig, ax = plt.subplots()
# the size of A4 paper
fig.set_size_inches(11.7, 8.27)

sns.barplot(x = 'STATION',y='TOTAL_TRAFFIC',data=data, ax=ax);
```
