```python
import pandas as pd
import glob
import numpy as np
import functools
import sys
import pickle
from datetime import date
```

```python
with open('cleanedMTAData','rb') as read_file:
    df_filter_months = pickle.load(read_file)
```

```python
df_station = df_filter_months.groupby((['STATION','DAY_OF_WEEK','WEEKFRAME','DATE']),sort=True).sum().reset_index()
```

```python
df_station[df_station['STATION'] == '34 ST-PENN STA'] # Sanity Check
```

# Average Weekday and Weekend Footfall Among Top 10 MTA Stations Pickle

```python
df_average_weekframe_group = df_station.groupby(['STATION', 'WEEKFRAME'], sort=True).mean().reset_index()
df_average_weekframe_group

df_average_weekday = df_average_weekframe_group[df_average_weekframe_group['WEEKFRAME'] == 'WEEK']
df_average_weekend = df_average_weekframe_group[df_average_weekframe_group['WEEKFRAME'] == 'WEEKEND']

df_average_weekday = df_average_weekday.sort_values(by = ['TOTAL_TRAFFIC'],ascending=[False])
df_average_weekend = df_average_weekend.sort_values(by = ['TOTAL_TRAFFIC'],ascending=[False])

# print(df_average_weekday)
# print(df_average_weekend)

df_average_weekdayTop10 = df_average_weekday[:10]
df_average_weekendTop10 = df_average_weekend[:10]

df_average_weekdayAndWeekend = df_average_weekdayTop10.merge(df_average_weekendTop10, how = 'outer')

print(df_average_weekdayAndWeekend)
# print(df_average_weekdayTop10)
# print(df_average_weekendTop10)

with open('top10weekdayAndWeekend_average', 'wb') as to_write:
    pickle.dump(df_average_weekdayAndWeekend, to_write)
```

# Total Weekday and Weekend Footfall Among Top 10 MTA Stations Pickle

```python
df_total_weekframe_group = df_station.groupby(['STATION', 'WEEKFRAME'], sort=True).sum().reset_index()
# weekdayAndWeekend = weekdayAndWeekend.groupby(['WEEKFRAME'], sort=True).sum().reset_index()
df_total_weekday = df_total_weekframe_group[df_total_weekframe_group['WEEKFRAME'] == 'WEEK']
df_total_weekend = df_total_weekframe_group[df_total_weekframe_group['WEEKFRAME'] == 'WEEKEND']

df_total_weekday = df_total_weekday.sort_values(by = ['TOTAL_TRAFFIC'],ascending=[False])
df_total_weekend = df_total_weekend.sort_values(by = ['TOTAL_TRAFFIC'],ascending=[False])

df_total_weekdayTop10 = df_total_weekday[:10]
df_total_weekendTop10 = df_total_weekend[:10]

df_total_weekdayAndWeekend = df_total_weekdayTop10.merge(df_total_weekendTop10, how = 'outer')

print(df_total_weekdayAndWeekend)
# print(df_total_weekdayTop10)
# print(df_total_weekendTop10)

with open('top10weekdayAndWeekend_total', 'wb') as to_write:
    pickle.dump(df_total_weekdayAndWeekend, to_write)
```
